```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import r2_score
import pickle
dataset=pd.read_csv("50_Startups.csv")
dataset = pd.get_dummies(dataset, drop_first=True)
dataset = dataset.astype(int)
X = dataset.drop(columns="Profit")
y = dataset[["Profit"]]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
regressor=SVR(kernel='rbf',C=1000)
regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)
r_score=r2_score(y_test,y_pred)
filename="finalized_model_svr.sav"
pickle.dump(regressor,open(filename,'wb'))
preinput=sc.transform([[1000,3000,3469,0,1]])
loded_model=pickle.load(open("finalized_model_svr.sav",'rb'))
result=loded_model.predict (preinput)
result
```

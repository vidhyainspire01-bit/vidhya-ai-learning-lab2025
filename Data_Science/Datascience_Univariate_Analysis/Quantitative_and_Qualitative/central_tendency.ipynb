```python
def quanqual(dataset):
    quan = dataset.select_dtypes(include=["number"]).columns.tolist()
    qual = dataset.select_dtypes(include=["object", "category", "bool"]).columns.tolist()
    stats_list={}
    for column in quan:
        stats_list[column]={
         
            "Mean": round(dataset[column].mean(), 2),
            "Median": round(dataset[column].median(), 2),
            "Mode": dataset[column].mode().iloc[0] if not dataset[column].mode().empty else None
        }
        stats_table = pd.DataFrame(stats_list).T.reset_index()
        stats_table.rename(columns={"index": "Numerical Column Name"}, inplace=True)
    return quan, qual, stats_table
quan,qual,stats_table =  quanqual(dataset)
#stats_table.to_csv("numerical_summary_1.csv", index=False)
print("\nStatistics for Quantitative Columns:")
print(stats_table)
```

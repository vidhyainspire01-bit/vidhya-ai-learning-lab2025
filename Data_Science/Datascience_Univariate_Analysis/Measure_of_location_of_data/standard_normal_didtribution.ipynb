```python
import pandas as pd
dataset=pd.read_csv("preplacement.csv")
dataset
```

```python
import seaborn as sns
def stdND(dataset):
    mean=dataset.mean()
    std=dataset.std()
    z_score = [(i - mean) / std for i in dataset]
    sns.histplot(z_score,kde=True)
   # return sum(z_score)/len(z_score)
    
```

```python
stdND(dataset["hsc_p"])
```

```python
def z_score(x, mean, std):
   
    return (x - mean) / std

john_z = z_score(2.85, 3.0, 0.7)
ali_z  = z_score(77, 80, 10)

print("John Z-score:", round(john_z, 2))
print("Ali Z-score:", round(ali_z, 2))
#John performed better relative to his school (closer to the mean) than Ali.
```

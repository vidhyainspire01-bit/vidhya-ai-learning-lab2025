```python
import seaborn as sns
import matplotlib.pyplot as plt  
from scipy.stats import norm
def check_distribution(dataset,startrange,endrange):
# Histogram (green) + KDE (blue)
    plt.figure(figsize=(7,4))
    ax=sns.histplot(dataset, kde=True, stat="density", bins=30, color="skyblue")
    sns.kdeplot(dataset, color='blue', ax=ax)
    
    # Vertical lines
    plt.axvline(startrange, color='red', linestyle="-")
    plt.axvline(endrange, color='red', linestyle="-")
    
    dataset_mean=dataset.mean()
    dataset_std=dataset.std()
    
    print("Mean=%.3f,Standard Deviation=%.3f" %(dataset_mean,dataset_std))
    dist=norm(dataset_mean,dataset_std)
    
    values=[value for value in range(startrange,endrange)]
           
    probabilities=[dist.pdf(value) for value in values]
    prob=sum(probabilities)
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return float(prob)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

def check_distribution_column(series, lower=60, upper=70):
    """Check distribution for a specific column (P between lowerâ€“upper)."""
    data = pd.to_numeric(series, errors="coerce").dropna()
    mu, sigma = data.mean(), data.std()

    # Probability from Normal Distribution
    prob = norm.cdf(upper, mu, sigma) - norm.cdf(lower, mu, sigma)

    # Empirical proportion (actual data between range)
    emp = ((data >= lower) & (data <= upper)).mean()

    # Plot distribution with shaded area
    plt.figure(figsize=(7,4))
    sns.histplot(data, kde=True, stat="density", color="skyblue", bins=30)

    # Shade the region
    x = np.linspace(data.min(), data.max(), 200)
    y = norm.pdf(x, mu, sigma)
    mask = (x >= lower) & (x <= upper)
    plt.fill_between(x[mask], y[mask], 0, alpha=0.3, color="blue")

    plt.axvline(lower, color="red", linestyle="--")
    plt.axvline(upper, color="red", linestyle="--")
    plt.title(f"Distribution of Column | Normal P={prob:.3f}, Empirical={emp:.3f}")
    plt.show()

    return {"Mean": mu, "Std": sigma, "NormalProb": prob, "EmpiricalProb": emp}
df = pd.read_csv("preplacement.csv")

# Pass only one column
result = check_distribution_column(df["ssc_p"], 60, 70)
print(result)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

def check_distribution_column(series, lower=60, upper=70):
    """Check distribution for a specific column (P between lowerâ€“upper)."""
    data = pd.to_numeric(series, errors="coerce").dropna()
    mu, sigma = data.mean(), data.std()

    # Probability from Normal Distribution
    prob = norm.cdf(upper, mu, sigma) - norm.cdf(lower, mu, sigma)

    # Empirical proportion (actual data between range)
    emp = ((data >= lower) & (data <= upper)).mean()

    # Plot distribution with shaded area
    plt.figure(figsize=(7,4))
    sns.histplot(data, kde=True, stat="density", color="skyblue", bins=30)

    # Shade the region
    x = np.linspace(data.min(), data.max(), 200)
    y = norm.pdf(x, mu, sigma)
    mask = (x >= lower) & (x <= upper)
    plt.fill_between(x[mask], y[mask], 0, alpha=0.3, color="blue")

    plt.axvline(lower, color="red", linestyle="-")
    plt.axvline(upper, color="red", linestyle="-")
    plt.title(f"Distribution | Normal P={prob:.3f}, Empirical={emp:.3f}")
    plt.show()

    # Convert to native Python floats
    return {
        "Mean": float(mu),
        "Std": float(sigma),
        "NormalProb": float(prob),
        "EmpiricalProb": float(emp)
    }
df = pd.read_csv("preplacement.csv")

result = check_distribution_column(df["ssc_p"], 50, 70)
print(result)
```
